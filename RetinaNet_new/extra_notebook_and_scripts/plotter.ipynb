```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the summary CSV
df = pd.read_csv("retinanet_evaluation_summary.csv")


# Create figure and axes
plt.figure(figsize=(12, 6))

# Plot each metric
plt.plot(df["Epoch"], df["(AP) @[ IoU=0.50:0.95"], label="AP @ IoU=0.50:0.95", marker='o')
plt.plot(df["Epoch"], df["(AP) @[ IoU=0.50"], label="AP @ IoU=0.50", marker='s')
plt.plot(df["Epoch"], df["(AP) @[ IoU=0.75"], label="AP @ IoU=0.75", marker='^')
plt.plot(df["Epoch"], df["(AR) @[ IoU=0.50:0.95"], label="AR @ IoU=0.50:0.95", marker='d')

# Title and labels
plt.title("RetinaNet Evaluation Metrics per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.xticks(df["Epoch"])  # Ensure all epochs appear on x-axis
plt.grid(True)

# Legend at the bottom
plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=False)

# Layout and save
plt.tight_layout()
plt.savefig("retinanet_metrics_plot.png", bbox_inches='tight')
plt.show()
display(df.head(20))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the summary CSV
file_name = "retinanet_depth_101_epoch_30_patience_1.csv"
df = pd.read_csv(file_name)

def plot_figure(file_name):
    # Create figure and axes
    plt.figure(figsize=(12, 6))

    # Plot each metric
    plt.plot(df["Epoch"], df["(AP) @[ IoU=0.50:0.95"], label="AP @ IoU=0.50:0.95", marker='o')
    plt.plot(df["Epoch"], df["(AP) @[ IoU=0.50"], label="AP @ IoU=0.50", marker='s')
    plt.plot(df["Epoch"], df["(AP) @[ IoU=0.75"], label="AP @ IoU=0.75", marker='^')
    plt.plot(df["Epoch"], df["(AR) @[ IoU=0.50:0.95"], label="AR @ IoU=0.50:0.95", marker='d')

    # Title and labels
    plt.title("RetinaNet Evaluation Metrics per Epoch")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.xticks(df["Epoch"])  # Ensure all epochs appear on x-axis
    plt.grid(True)

    # Legend at the bottom
    plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=False)

    # Layout and save
    plt.tight_layout()
    plt.savefig(f"{file_name.split(".")[0]}.png", bbox_inches='tight')
    plt.show()

plot_figure(file_name)
display(df.head(20))
```
